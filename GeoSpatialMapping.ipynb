```python
import pandas as pd
import numpy as np
import sqlite3 as db
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# https://www.dataquest.io/blog/python-pandas-databases/
```

```python
tbl1= pd.read_csv("AHS_districtwise_2012-13.csv")
tbl1.head()
```

```python
if 'RURAL'  in 'State'.upper():
    print("Rural")
elif "URBAN" in 'State'.upper():
    print("Urban")
elif "TOTAL" in 'State'.upper():
    print("Total")
else:
    print("Other")
```

```python
col_names=tbl1.columns
```

```python
#Find out Which columns have urban/rural in them
col_desc=[]
for col_name in col_names:
    if 'RURAL'  in col_name.upper():
        col_desc.append("Rural")
    elif "URBAN" in col_name.upper():
        col_desc.append("Urban")
    elif "TOTAL" in col_name.upper():
        col_desc.append("Total")
    else:
        col_desc.append("Other")

print(col_desc, "\n\n", "Total no of columns =",len(col_desc))
```

```python
col_info=pd.DataFrame({"col_names":col_names, "col_desc":col_desc}, index=None)
col_info.head()
col_info.to_csv("column_info.csv", index=False)
```

```python
col_info= pd.read_csv("column_info.csv")
col_info.head()
```

```python
col_Total = col_info[col_info["col_desc"].isin(["Other", "Total"])]
col_Rural = col_info[col_info["col_desc"].isin(["Other", "Rural"])]
col_Urban = col_info[col_info["col_desc"].isin(["Other", "Urban"])]
print(col_Total.head(10),"\n")
print(col_Rural.head(10),"\n")
print(col_Urban.head(10),"\n")
```

```python
tbl_Total = tbl1[tbl1.columns.intersection(col_Total["col_names"])]
tbl_Urban = tbl1[tbl1.columns.intersection(col_Urban["col_names"])]
tbl_Rural = tbl1[tbl1.columns.intersection(col_Rural["col_names"])]

print(tbl_Total.head(), "\n\n")
print(tbl_Urban.head(), "\n\n")
print(tbl_Rural.head(), "\n\n")
```

```python
tbl_Total.to_csv("Total.csv", index = False)
tbl_Urban.to_csv("Urban.csv", index = False)
tbl_Rural.to_csv("Rural.csv", index = False)
```

```python
import re
stats_needed = []
for i in col_Total["col_names"]:
    if "-" in i:
        stats = re.search("-(.*)-",i)
        stats_needed.append(stats.group(1))
    else:
        continue

stats_needed_tbl = pd.DataFrame(stats_needed)
stats_needed_tbl.to_csv("stats_needed.csv", header=True, index= False)        
print(stats_needed)
```

```python
districts = tbl1[["State", "State / District Name"]]
districts= districts[districts["State"] != districts["State / District Name"]]
districts.to_csv("Districts.csv", header=True, index= False)
```

# Missing Value Treatment

```python
# tbl_Total_num = tbl_Total.select_dtypes(include=np.number)
for col in range(2,216):
    for row in range(293):
        if tbl_Total.iloc[row,col] == 0 or pd.isnull(tbl_Total.iloc[row,col]):
            tbl_Total.iloc[row,col] = tbl_Total.iloc[:,col].mean()
tbl_Total.head()
```

```python
# QC
tbl_Total_null = tbl_Total[tbl_Total.isnull().any(axis=1)]
tbl_Total_null
```
